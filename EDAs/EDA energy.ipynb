```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
```

```python
sns.set(style='whitegrid', palette='crest')
```

```python
energia=pd.read_csv('energiausa.csv')
```

```python
energia.head()
```

```python
energia.info()
```

```python
energia.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
```

```python
energia.describe()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Energy_type', y='Energy_consumption', data=energia, palette='crest')
plt.title('Distribución de Consumo de Energía por Tipo de Energía')
plt.xlabel('Tipo de Energía')
plt.ylabel('Consumo de Energía')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Energy_production', y='Energy_consumption', hue='Year', data=energia, palette='crest', alpha=0.6)
plt.title('Producción vs Consumo de Energía')
plt.xlabel('Producción de Energía')
plt.ylabel('Consumo de Energía')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='GDP', y='CO2_emission', hue='Year', data=energia, palette='crest', alpha=0.6)
plt.title('Emisiones de CO2 vs PIB')
plt.xlabel('PIB')
plt.ylabel('Emisiones de CO2')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='Energy_production', hue='Energy_type', data=energia, palette='crest')
plt.title('Energía Producida por Año')
plt.xlabel('Año')
plt.ylabel('Energía Producida')
plt.legend(title='Tipo de Energía', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='Energy_consumption', hue='Energy_type', data=energia, palette='crest', linestyle='--')
plt.title('Energía Consumida por Año')
plt.xlabel('Año')
plt.ylabel('Energía Consumida')
plt.legend(title='Tipo de Energía', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Energy_type', y='Energy_intensity_per_capita', data=energia, palette='crest')
plt.title('Intensidad Energética per Cápita por Tipo de Energía')
plt.xlabel('Tipo de Energía')
plt.ylabel('Intensidad Energética per Cápita')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Energy_type', y='Energy_intensity_by_GDP', data=energia, palette='crest')
plt.title('Intensidad Energética por PIB por Tipo de Energía')
plt.xlabel('Tipo de Energía')
plt.ylabel('Intensidad Energética por PIB')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='CO2_emission', hue='Energy_type', data=energia, palette='crest')
plt.title('Emisiones de CO2 a lo Largo del Tiempo por Tipo de Energía')
plt.xlabel('Año')
plt.ylabel('Emisiones de CO2')
plt.legend(title='Tipo de Energía', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
energia_numerica = energia.select_dtypes(include=[float, int])

# Calcular la matriz de correlación
correlation_matrix = energia_numerica.corr()

# Crear el heatmap con la matriz de correlación
plt.figure(figsize=(14, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='crest')
plt.title('Matriz de Correlación entre Variables')
plt.show()

```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Population', y='Energy_consumption', hue='Energy_type', data=energia, palette='crest', alpha=0.6)
plt.title('Relación entre Población y Consumo de Energía')
plt.xlabel('Población')
plt.ylabel('Consumo de Energía')
plt.legend(title='Tipo de Energía', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='GDP', y='Energy_consumption', hue='Energy_type', data=energia, palette='crest', alpha=0.6)
plt.title('Relación entre PIB y Consumo de Energía')
plt.xlabel('PIB')
plt.ylabel('Consumo de Energía')
plt.legend(title='Tipo de Energía');
```

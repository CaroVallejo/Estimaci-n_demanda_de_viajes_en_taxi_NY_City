```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
```

```python
sns.set(style='whitegrid', palette='crest')
```

```python
energia=pd.read_csv('energiausa.csv')
```

```python
energia.head()
```

```python
energia.info()
```

```python
energia.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
```

```python
energia.describe()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Energy_type', y='Energy_consumption', data=energia, palette='crest')
plt.title('Distribucin de Consumo de Energa por Tipo de Energa')
plt.xlabel('Tipo de Energa')
plt.ylabel('Consumo de Energa')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Energy_production', y='Energy_consumption', hue='Year', data=energia, palette='crest', alpha=0.6)
plt.title('Produccin vs Consumo de Energa')
plt.xlabel('Produccin de Energa')
plt.ylabel('Consumo de Energa')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='GDP', y='CO2_emission', hue='Year', data=energia, palette='crest', alpha=0.6)
plt.title('Emisiones de CO2 vs PIB')
plt.xlabel('PIB')
plt.ylabel('Emisiones de CO2')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='Energy_production', hue='Energy_type', data=energia, palette='crest')
plt.title('Energa Producida por Ao')
plt.xlabel('Ao')
plt.ylabel('Energa Producida')
plt.legend(title='Tipo de Energa', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='Energy_consumption', hue='Energy_type', data=energia, palette='crest', linestyle='--')
plt.title('Energa Consumida por Ao')
plt.xlabel('Ao')
plt.ylabel('Energa Consumida')
plt.legend(title='Tipo de Energa', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Energy_type', y='Energy_intensity_per_capita', data=energia, palette='crest')
plt.title('Intensidad Energtica per Cpita por Tipo de Energa')
plt.xlabel('Tipo de Energa')
plt.ylabel('Intensidad Energtica per Cpita')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Energy_type', y='Energy_intensity_by_GDP', data=energia, palette='crest')
plt.title('Intensidad Energtica por PIB por Tipo de Energa')
plt.xlabel('Tipo de Energa')
plt.ylabel('Intensidad Energtica por PIB')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='CO2_emission', hue='Energy_type', data=energia, palette='crest')
plt.title('Emisiones de CO2 a lo Largo del Tiempo por Tipo de Energa')
plt.xlabel('Ao')
plt.ylabel('Emisiones de CO2')
plt.legend(title='Tipo de Energa', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
energia_numerica = energia.select_dtypes(include=[float, int])

# Calcular la matriz de correlacin
correlation_matrix = energia_numerica.corr()

# Crear el heatmap con la matriz de correlacin
plt.figure(figsize=(14, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='crest')
plt.title('Matriz de Correlacin entre Variables')
plt.show()

```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Population', y='Energy_consumption', hue='Energy_type', data=energia, palette='crest', alpha=0.6)
plt.title('Relacin entre Poblacin y Consumo de Energa')
plt.xlabel('Poblacin')
plt.ylabel('Consumo de Energa')
plt.legend(title='Tipo de Energa', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='GDP', y='Energy_consumption', hue='Energy_type', data=energia, palette='crest', alpha=0.6)
plt.title('Relacin entre PIB y Consumo de Energa')
plt.xlabel('PIB')
plt.ylabel('Consumo de Energa')
plt.legend(title='Tipo de Energa');
```
